```python
import pandas as pd
```

```python
df = pd.read_csv('../data/PEDE_PASSOS_DATASET_FIAP.csv', sep=';', encoding='utf-8')
```

```python
df.iloc[[1258]]
```

```python
df = df.drop([1258], axis=0).reset_index(drop=True)
```

```python
df_2020 = df[['IAN_2020',
              'IDA_2020',
              'IEG_2020',
              'IAA_2020',
              'IPS_2020',
              'IPP_2020',
              'IPV_2020']]

df_2021 = df[['IAN_2021',
              'IDA_2021',
              'IEG_2021',
              'IAA_2021',
              'IPS_2021',
              'IPP_2021',
              'IPV_2021']]

df_2022 = df[['IAN_2022',
              'IDA_2022',
              'IEG_2022',
              'IAA_2022',
              'IPS_2022',
              'IPP_2022',
              'IPV_2022']]
```

```python
def rename_columns(df, ano):
    df.rename(columns={f'IAN_{ano}': 'IAN',
                       f'IDA_{ano}': 'IDA',
                       f'IEG_{ano}': 'IEG',
                       f'IAA_{ano}': 'IAA',
                       f'IPS_{ano}': 'IPS',
                       f'IPP_{ano}': 'IPP',
                       f'IPV_{ano}': 'IPV'}, inplace=True)
```

```python
rename_columns(df_2020, '2020')
rename_columns(df_2021, '2021')
rename_columns(df_2022, '2022')
```

```python
df_2020['Ano'] = '2020'
df_2021['Ano'] = '2021'
df_2022['Ano'] = '2022'
```

```python
df_indicadores = pd.concat([df_2020, df_2021, df_2022], axis=0, ignore_index=True)
```

```python
df_indicadores.to_csv('../data/Indicadores.csv', sep=';', index=False)
```
